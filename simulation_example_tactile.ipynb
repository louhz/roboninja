```python
import os
import numpy as np
import matplotlib.pyplot as plt

from hydra import compose, initialize_config_dir
from hydra.core.global_hydra import GlobalHydra

from roboninja.env.tc_env import get_strawberry_cuthalf_env_tactile
from roboninja.utils.misc import animate

import pickle

def forward(
    taichi_env,
    init_state,
    comp_actions,
    grad_enabled,
    render,
    # --- NEW ---
    use_tactile=True,
    print_tactile=True,
    print_only_during_cut=True,
):
    taichi_env.set_state(init_state, grad_enabled=grad_enabled)

    actions_v = comp_actions[:-1]
    action_p = comp_actions[-1]
    horizon_action = len(actions_v)

    # apply the "p" action (pose/param)
    taichi_env.apply_agent_action_p(action_p)

    render_images = []
    tactile_log = []  # NEW: store tactile readings per action step

    if render:
        render_images.append(taichi_env.render())

    # Helper: define "during cut" as when your velocity has negative Y (your script uses v[:,1] = -0.3)
    def _is_cut_step(v):
        return v[1] < 0.0

    # Helper: read tactile in a compatible way
    def _step_and_get_tactile(v):
        """
        Prefers env.step(..., return_tactile=True) if available.
        Falls back to env.get_tactile() after step.
        """
        # Your TaichiEnvtactile.step supports return_tactile
        try:
            return taichi_env.step(v, return_tactile=True)
        except TypeError:
            # If step signature doesn't have return_tactile, do it in two calls
            taichi_env.step(v)
            if hasattr(taichi_env, "get_tactile"):
                return taichi_env.get_tactile()
            return {"contact": 0, "pulse": 0.0, "count": 0}

    # Rollout
    for i in range(horizon_action):
        v = actions_v[i]

        tactile = None
        if use_tactile and hasattr(taichi_env, "get_tactile"):
            tactile = _step_and_get_tactile(v)
            tactile_log.append(tactile)

            do_print = print_tactile and (not print_only_during_cut or _is_cut_step(v))
            if do_print:
                # tactile is documented as: {"contact": int, "pulse": float, "count": int}
                contact = tactile.get("contact", 0)
                pulse = tactile.get("pulse", 0.0)
                count = tactile.get("count", 0)
                print(f"[step {i:03d}] tactile: contact={contact}  count={count}  pulse={pulse:.6f}")

        else:
            taichi_env.step(v)

        if render:
            render_images.append(taichi_env.render())

    loss = taichi_env.get_loss()

    return {
        "loss": loss,
        "render_images": render_images,
        "tactile_log": tactile_log,  # NEW
    }


def backward(taichi_env, comp_actions, **kwargs):
    taichi_env.reset_grad()
    taichi_env.get_loss_grad()

    actions_v = comp_actions[:-1]
    horizon_action = len(actions_v)

    for i in range(horizon_action - 1, -1, -1):
        taichi_env.step_grad(actions_v[i])

    grad = taichi_env.agent.get_grad(len(actions_v))
    grad[-1][1:] = 0

    return {"grad": grad}


# ---------------- Hydra setup (same as your script) ----------------
if GlobalHydra.instance().is_initialized():
    GlobalHydra.instance().clear()
GlobalHydra.instance().clear()

config_dir = os.path.abspath("roboninja/config/cut_env")
initialize_config_dir(config_dir=config_dir, version_base=None)

cfg = compose(config_name="cut_env_strawberry")
bone_idx = 0

cfg.strawberry.mesh_path = "/home/louhz/Desktop/Rss/roboninja/generateddata/Generated_strawberry/0001/mesh.obj"
cfg.strawberry.pos = (0.5, 0.1, 0.45)  # (x,z,y) in your comment
cfg.strawberry.scale = (0.25, 0.25, 0.25)

cfg.strawberry.voxelize_res = 256
cfg.strawberry.shell_radius_vox = 2
cfg.strawberry.close_iters = 2
cfg.strawberry.normalize = True
cfg.strawberry.trim_percentile = 0.5
cfg.strawberry.cache_voxels = True
cfg.strawberry.euler = (0.0, 180.0, 0.0)
cfg.auto_boundary = True

cfg.bone.name = f"bone_{bone_idx}"

# ---------------- NEW: tactile sensor config ----------------
# IMPORTANT:
# - TaichiEnvtactile/MPMSimulatortactile requires setup_tactile_sensor BEFORE env.build().
# - If your get_strawberry_cuthalf_env() builds the env internally (common),
#   then you MUST pass tactile_sensor_cfg into the env factory and have it forward
#   into TaichiEnvtactile(..., tactile_sensor_cfg=...).
#
# If you already updated get_strawberry_cuthalf_env to accept tactile_sensor_cfg, this will work:
tactile_sensor_cfg = dict(
    # TODO: set these to match your knife patch / coordinate convention.
    # Many implementations interpret this in the KNIFE LOCAL frame (recommended for a knife-mounted sensor),
    # but some interpret in WORLD frame. Use whatever your MPMSimulatortactile expects.
    center=(0.5, 0.1, 0.45),
    length=0.06,
    height=0.02,
    normal_axis=1,
    thickness=0.002,
    pulse_on_entry=True,
    one_sided=False,
    front_side="positive",
    record_per_particle=False,
)

try:
    taichi_env = get_strawberry_cuthalf_env_tactile(cfg, tactile_sensor_cfg=tactile_sensor_cfg)
except TypeError:
    # If your env factory doesn't accept tactile_sensor_cfg yet,
    # you need to add it there (recommended).
    taichi_env = get_strawberry_cuthalf_env_tactile(cfg)

# Sanity print
if hasattr(taichi_env, "has_tactile_sensor"):
    print("has_tactile_sensor:", taichi_env.has_tactile_sensor)

knife = taichi_env.agent.effectors[0]

# ---------------- Rollout params (same as your script) ----------------
horizon_action = 60

init_action_p = np.asarray([-0.6, 0.155, 0.5])
init_action_v = np.asarray([[0.0, 0.0, 0.0]] * horizon_action)
init_action_v[: int(horizon_action * 0.4), 1] = -0.3  # "cut phase" (negative Y velocity)
init_actions = np.concatenate([init_action_v, init_action_p[None]], axis=0)

init_state = taichi_env.get_state()["state"]
current_actions = init_actions

kwargs = dict(
    taichi_env=taichi_env,
    init_state=init_state,
    comp_actions=current_actions,
    grad_enabled=True,
    render=True,
)

# ---------------- Run forward with tactile printing ----------------
out = forward(
    **kwargs,
    use_tactile=True,
    print_tactile=True,
    print_only_during_cut=True,  # prints tactile only when v[1] < 0
)

print("loss:", out["loss"])

# Optional: animate renders
animate(out["render_images"])

# Optional: quick plot of tactile 'count' over time (if you want)
if out["tactile_log"]:
    tactile_count = [t.get("count", 0) for t in out["tactile_log"]]
    plt.figure()
    plt.plot(tactile_count)
    plt.title("Tactile count per action step")
    plt.xlabel("action step")
    plt.ylabel("count")
    plt.show()
```
